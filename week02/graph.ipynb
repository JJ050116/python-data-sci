```python
import matplotlib.pyplot as plt
```

```python
players = ['Messi', 'Ronaldo', 'Haaland', 'Neymar Jr', 'Mbappe']
goals = [100, 101, 89, 73, 81]

plt.bar(players, goals)

plt.title('Goal this season')
plt.xlabel('Players')
plt.ylabel('Goals')

plt.show()
```

```python
snacks = ['Pizza', 'Fries', 'Ice cream', 'Pie']
likes = [9, 8, 10, 4]

plt.bar(snacks, likes, color='purple')
plt.title('My fav snacks')
plt.show()
```
